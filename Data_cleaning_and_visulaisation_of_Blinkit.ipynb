```python
import pandas as pd
file_path = '/content/drive/My Drive/Copy of BlinkIT Grocery Data.xlsx'
df = pd.read_excel(file_path)
df.head()
```

```python


# Preview
print("🔍 Data Preview:")
display(df.head())

# Data Info
print("\n📦 Dataset Info:")
df.info()

# Check for missing values
print("\n🧼 Null Values:")
print(df.isnull().sum())

# Drop duplicates
df.drop_duplicates(inplace=True)

# Fill missing numerical columns with mean
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].mean(), inplace=True)

# Fill missing categorical columns with mode
for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)

# Show cleaned data preview
print("\n✅ Cleaned Data:")
display(df.head())
```

```python
import plotly.express as px

# 1. 🧈 Fat Content Distribution
fig1 = px.histogram(df, x="Item Fat Content", color="Item Fat Content", title="🧈 Fat Content Distribution")
fig1.show()
# Top 10 Item Types
top_types = df['Item Type'].value_counts().nlargest(10)
top_df = top_types.reset_index()
top_df.columns = ['Item Type', 'Count']  # Rename columns correctly

import plotly.express as px

fig2 = px.bar(
    top_df,
    x='Item Type', y='Count',
    color='Item Type',
    title="🍽️ Top 10 Item Types"
)
fig2.show()


# 3. 🏪 Outlet Type Distribution
fig3 = px.pie(df, names='Outlet Type', title="🏪 Outlet Type Distribution", hole=0.4)
fig3.show()

# 4. 📦 Sales by Outlet Type
fig4 = px.box(df, x='Outlet Type', y='Sales', color='Outlet Type',
              title="📦 Sales by Outlet Type")
fig4.show()

# 5. 🔍 Item Visibility vs Sales Scatter (Interactive)
fig5 = px.scatter(df, x='Item Visibility', y='Sales', color='Item Fat Content',
                  title="🔍 Sales vs Item Visibility", opacity=0.5,
                  hover_data=['Item Type', 'Outlet Type'])
fig5.show()

# 6. ⭐ Rating Distribution
fig6 = px.histogram(df, x="Rating", nbins=15, title="⭐ Rating Distribution")
fig6.show()
```
